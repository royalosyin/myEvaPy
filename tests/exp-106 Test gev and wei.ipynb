```python
import numpy as np
import lmoments3 as lm3
from lmoments3 import distr
from scipy.stats import genextreme, weibull_min
from scipy.special import comb as scomb, gammaln as sgammaln
import rvlib as rl

from gev_dist import gev_lmom_fit, gev_ppf, gev_rvs
from wei_dist import wei_lmom_fit, wei_ppf, wei_bs_reps_ci, wei_rvs
```

```python
data = [2.0, 3.0, 4.0, 2.4, 5.5, 1.2, 5.4, 2.2, 7.1, 1.3, 1.5, 12.1, 9.8, 15.3, 25, 112, 23, 54, 56, 75, 11, 33]
avis  = np.array([2, 3, 5, 10, 15, 20, 50, 100, 200, 300, 500, 1000, 10000])
probs = 1.0 - 1.0/avis
```

## GEV

```python
xparas = distr.gev.lmom_fit(data)
print(xparas)
fitted_gam = distr.gev(**xparas)
fitted_gam.ppf(probs)
```

```python
u, l, s = gev_lmom_fit(np.array(data))
print(u, l, s)
gev_ppf(probs, u, l, s)
```

## Wei

```python
#%%timeit
wparas = distr.wei.lmom_fit(data)
#print(wparas)
fitted_wei = distr.wei(**wparas)
fitted_wei.ppf(probs)
```

```python
#%%timeit
u, l, s = wei_lmom_fit(np.array(data))
#print(u, l, s)
wei_ppf(probs, u, l, s)
```
