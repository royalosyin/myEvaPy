```python
import numpy as np
import xarray as xr
import scipy.stats as stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
in_file = 'data/PRATE_surface.cmd_mm_9mon.nc'
da_rain = xr.open_dataset(in_file)['PRATE_surface']
#da_rain
nt, nrow, ncol = da_rain.shape 
```

```python
%%time
for irow in np.arange(nrow):
    for icol in np.arange(ncol):
        data = da_rain[:,irow, icol]
        data = data[data>0]*24*3600*30
        it_alpha, fit_loc, fit_beta=stats.gamma.fit(data, floc=0)
        #print(it_alpha, fit_beta)
```
